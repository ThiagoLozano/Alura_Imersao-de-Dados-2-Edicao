```
# Importa a Biblioteca Pandas.
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Pega a fonte dos dados.
fonte = "https://github.com/alura-cursos/imersao-dados-2-2020/blob/master/MICRODADOS_ENEM_2019_SAMPLE_43278.csv?raw=true"

# Cria o objeto 'dados'.
dados = pd.read_csv(fonte)
```

```
# Cria uma lista com todas as Provas.
provas = ["NU_NOTA_CN", "NU_NOTA_CH", "NU_NOTA_MT", "NU_NOTA_LC", "NU_NOTA_REDACAO"]
```

```
# Cria uma coluna para Nota Total das Provas.
dados['NU_NOTA_TOTAL'] = dados[provas].sum(axis=1)
```

```
# Retorna alunos que não tem o total de notas igual a 0.
dados_sem_nota_zero = dados.query('NU_NOTA_TOTAL != 0')
```

```
# Retorna um gráfico de Proporção de Inscritos Por Nota Total.
plt.figure(figsize=(12,8))
sns.histplot(dados_sem_nota_zero, x = 'NU_NOTA_TOTAL')
```

```
# Retorna um gráfico de Proporção de Inscritos Pela nota de Matemática.
plt.figure(figsize=(12,8))
sns.histplot(dados_sem_nota_zero, x = 'NU_NOTA_MT')
```

```
# Retorna um gráfico de Proporção de Inscritos pela nota de Liguagens e Códigos.
plt.figure(figsize=(12,8))
sns.histplot(dados_sem_nota_zero, x = 'NU_NOTA_LC')
```

```
# Retorna um gráfico de Proporção de Inscritos 
# B - Pessoas que tem internet
# A - Pessoas que não tem internet
plt.figure(figsize=(12,8))
sns.histplot(dados_sem_nota_zero, x = 'NU_NOTA_TOTAL', hue='Q025', kde=True)
```

```
# Mostra um gráfico de Disperção entre as notas de Matemática e Português.
sns.scatterplot(data=dados_sem_nota_zero, x='NU_NOTA_MT', y='NU_NOTA_LC')
plt.figure(figsize=(10,10))
plt.xlim((-50,1050))
plt.ylim((-50,1050))
```

```
# Mostra gráficos de Disperção entre todas as notas.
sns.pairplot(dados_sem_nota_zero[provas])
```

```
# Mostra a correlação.
# -1: Correlações Negativas
# +1: Correlações Postivas
# 0: Poucas correlações.
correlacao = dados_sem_nota_zero[provas].corr()
correlacao
```

```
# Mostra um gráfico de calor.
sns.heatmap(correlacao, cmap='Blues', center=0, annot=True)
```
