```
# Importa as Bibliotecas: Pandas, Matplotlib e Seaborn.
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Fonte dos dados
fonte = "https://github.com/alura-cursos/imersao-dados-2-2020/blob/master/MICRODADOS_ENEM_2019_SAMPLE_43278.csv?raw=true"

# Cria um objeto para a leitura do DataFrame.
dados = pd.read_csv(fonte)
```

## Relações Iniciais

```
# PERCENTAGEM DE HOMENS E MULHERES QUE SE INSCREVERAM #

# Pega a frequência de cada sexo.
frequencia_sexo = dados['TP_SEXO'].value_counts()

# Cria o gráfico para a Percentagem em Pizza.
plt.figure(figsize=(8,6))
plt.title('Percentagem de Homens e Mulheres que se inscreveram', fontdict={'fontsize':20})
frequencia_sexo.plot.pie(fontsize=15, autopct='%1.1f%%')

# Cria o gráfico para a Quantidade em Barras.
plt.figure(figsize=(8,6))
plt.title('Proporção de Homens e Mulheres que se inscreveram', fontdict={'fontsize':20})
plt.xlabel('Sexo', fontdict={'fontsize':20})
plt.ylabel('Quantidade de Inscritos', fontdict={'fontsize':20})
frequencia_sexo.plot.bar(fontsize=15, grid=True)
```

Aqui, podemos obsevar que a quantidade de inscritos é bem mais presente no sexo **Feminino**.

No primeiro gráfico podemos ver que a porcentagem do sexo **Feminino** tem uma diferença de **19.2%** em relação ao **Masculino**.

E no segundo gráfico a quantidade de inscritos **Mulheres** passa dos **70.000** enquando os inscritos **Homens** chega a passar bem pouco de **50.000.**

*Conclução:* Mesmo sendo uma amostra de dados, podemos perceber que as mulheres tem um grande espaço na sociedade e que elas representam aproximadamante um pouco mais de 50% dos estudantes que querem fazer o ENEM e conseguirem uma boa oportunidade acadêmica.

```
# PROPORÇÃO DE INSCRITOS REFERENTE AOS ESTADOS. #

# Pega a frequência de cada Estado Residencial.
frequencia_estados = dados['SG_UF_RESIDENCIA'].value_counts()

# Pega a percentagem de cada Estado Residencial.
percentual_estados = dados['SG_UF_RESIDENCIA'].value_counts(normalize=True)

# Cria o gráfico para a Quantidade em Barras.
plt.figure(figsize=(15,6))
plt.title('Proporção de Inscritos de cada Estado Residencial', fontdict={'fontsize':20})
plt.xlabel('Estados', fontdict={'fontsize':20})
plt.ylabel('Quantidade de Inscritos', fontdict={'fontsize':20})
frequencia_estados.plot.bar(fontsize=15, grid=True)

# Cria o gráfico para a Percentagem em Barras.
plt.figure(figsize=(15,6))
plt.title('Percentual de Inscritos de cada Estado Residencial', fontdict={'fontsize':20})
plt.xlabel('Estados', fontdict={'fontsize':20})
plt.ylabel('Percentual', fontdict={'fontsize':20})
percentual_estados.plot.bar(fontsize=15, grid=True, color='Green')
```

Aqui, podemos observar que **São Paulo** é o estado que tem mais inscritos e **Roraima** a que tem menos.

Algo que pode justificar isso é justamente a quantididade populacional de cada estado, que, segundo a **Wikipedia**, **São Paulo** se encontra na 1º Posição com aproximadamente **46.289.333** pessoas e Roraima está na 27º Posição sendo o último com **631.181** pessoas.
Já **Bahia** e **Rio de Janeiro**, possuem apenas uma diferença de **2.435.555** pessoas, tendo o Rio de Janeiro **17.366.189** e Bahia **14.930.634** o que não justifica muito o porque de **BH** ter mais inscritos do que **RJ**, talvez uma análise dos registros completos responda a isso.

A segunda tabela nos mostra a **porcentagem** que cada estado representa dentre do total. Como podemos ver, **SP** representa aproximadamente **16%** dos inscritos, seguido por **MG** com aproximadamente **11%** e **BH** com apoximadamente **8%**.

*Fonte:* https://pt.wikipedia.org/wiki/Lista_de_unidades_federativas_do_Brasil_por_popula%C3%A7%C3%A3o

*Conclusão:* A quantidade populacional está correlacionada ao estado com mais ou com menos inscritos, ou seja, quanto maior a população de um estado, maior a proporção de inscritos dela ou o inverso.

```
# PROPORÇÃO POR IDADE #

# Pega a Frequência de cada Idade.
frequencia_idade = dados['NU_IDADE'].value_counts().sort_index()
frequencia_dezoito_menos = dados.query("NU_IDADE <= 18")['NU_IDADE'].value_counts().sort_index()
frequencia_mais_setenta = dados.query("NU_IDADE > 70")['NU_IDADE'].value_counts().sort_index()

# Cria gráfico para a Proporção de Inscritos por Idade.
plt.figure(figsize=(20,6))
plt.title('Proporção de Inscritos por sua Idade', fontdict={'fontsize':20})
plt.xlabel('Idade', fontdict={'fontsize':20})
plt.ylabel('Quantidade de Inscritos', fontdict={'fontsize':20})
frequencia_idade.plot.bar(fontsize=15, grid=True, color='purple')

# Cria gráfico para a Proporção de Inscritos com 18 anos ou menos.
plt.figure(figsize=(10,6))
plt.title('Proporção de Inscritos com 18 anos ou menos', fontdict={'fontsize':20})
plt.xlabel('Quantidade de Inscritos', fontdict={'fontsize':20})
plt.ylabel('Idade', fontdict={'fontsize':20})
frequencia_dezoito_menos.plot.barh(fontsize=15, grid=True)

# Cria gráfico para a Proporção de Inscritos com mais de 70 anos.
plt.figure(figsize=(10,6))
plt.title('Proporção de Inscritos com mais de 70 anos', fontdict={'fontsize':20})
plt.xlabel('Idade', fontdict={'fontsize':20})
plt.ylabel('Quantidade de Inscritos', fontdict={'fontsize':20})
frequencia_mais_setenta.plot.bar(fontsize=15, grid=True, color='orange')
```

Aqui, podemos observar que a grande maioria dos participantes está na faixa etária entre **17** e **18** anos, o que era de se esperar, pois, esse é o momento onde os alunos **terminaram o Ensino Médio**.

Porém, o intuido do segundo gráfico foi mostrar que pessoas com menos de 18 anos também se inscreveram e o mais curioso, temos registros de pessoas entre **13** e **14** anos, praticamente crianças que nem se quer chegaram a entrar no ensino médio.

O terceiro gráfico mostra o 'outro lado da moeda', inscritos que já passam de **70 anos**, e mesmo sendo uma amostra, é muito interessante saber que há pessoas de idade que ainda pensam em conseguir um lugar no mundo acadêmico, quebrando preconceitos em relação a sua idade.
